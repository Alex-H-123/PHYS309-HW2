Recall that in PHYS 207, you were often asked to plot displacement as a function of time. For example, for a baseball thrown directly upward you would plot $y(t)$, and for a block on a horizontal spring you would plot $x(t)$. Here we will examine how the displacement and time axes "scissor together" at relativistic speeds.

The convention in relativity is to reverse the axes: $ct$ is vertical and $x$ is horizontal. Time moves forward from the bottom to the top of the plot:

Imagine that frame $K^{\prime}$ is moving in the $+x$ direction at speed $v$ relative to frame $K$. You are at rest in frame $K$. Let's figure out how the set of $(x^{\prime}, ct^{\prime})$ axes would look to you when the frames' origins $O$ and $O^{\prime}$ coincide.

#math used to find ct(x) that traces the ct' axis

using the lorentz transform and the given information . . .

x' = gamma(x-vt), x' =0

we find that x-vt = 0, since gamma is never 0

solving for t we find . . .

t = x/v, we need ct(x), so multiply both sides by c,
we find that ct(x) = cx/v

*I do not quite understand how we know that this ct(x) traces ct', but it made sense to solve the lorentz when x' = 0, I also found an article describing how the t' axis for the moving frame is 1/v, so this makes some sense.


```python
def ctprime_axis(v,x):
    ctprime = (constants.speed_of_light * x) / v
    return(ctprime)
```

#math used to find ct(x) that traces x'

Using the Lorentz and givne information . . .

t' = gamma(t - vx/c^2), ct' = 0

we need to find t - vx/c^2, as gamma can never be zero

solving for ct(x) we find . . .

t = vx/c^2

ct(x) = vx/c

```python
def xprime_axis(v,x):
    xprime = (v*x) / (constants.speed_of_light)
    return(xprime)
```
