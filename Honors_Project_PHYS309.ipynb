<h3> A review of quantum numbers </h3>

During the last portion of this course we learned about how to use spherical coordinates to obtain the time-independent Schrodinger wavefunction for the hydrogen atom. This wavefunctions can be used to describe the shape of the probability density for the hydrogen atom. Using seperation of variables we derived three different components of the wavefunction $\varphi(r, \theta, \phi)$.

By seperating this equation into three sperate parts R(r), f($\theta$), g($\phi$), we obtained three seperate equations:

$\bullet$ (The Azimuthal Equation) $\frac{d^2g}{d\phi^2} = -m_l^2g$

$\bullet$ (The Radial Equation) $\frac{1}{r^2}\frac{d}{d\theta}(r^2\frac{dR}{dr}) + \frac{2\mu}{\hbar^2}(E - V - \frac{\hbar^2}{2\mu}\frac{l(l+1)}{r^2})R = 0$

$\bullet$ (Angular Equation) $\frac{1}{\sin(\theta)}\frac{d}{d\theta}(\sin(\theta)\frac{df}{d\theta})+(l(l+1) - \frac{m_l^2}{\sin(\theta)^2})f = 0$

Without going into the full derivations (as we have focsued on this in class), these equations introduce the three quantum numbers $n$, $l$, and $m_l$. This is easily apparent for $l$ and $m_l$, however $n$ does not appear in any of these equations. $n$ is found from the Radial equation by inserting the Coloumb potential, $V = \frac{e^2}{4\pi\epsilon_0r}$, into the expression. When this is done it can be found that $E_n = -\frac{E_0}{n^2}$, where $E_0 = -13.6 eV$. Now that we have found the quantum numbers, what physical quantities do they relate too?

$\bullet$ The principle number ($n = 1,2,3,...$): relates to the energy state of the electron by the expression $E_n = -\frac{E_0}{n^2}$

$\bullet$ The orbital angular momentum quantum number ($l =  0,1,2,3,...,n-1 $): relates to the angular momentum of the electron-proton system by $L=\sqrt{l(l+1)}\hbar$

$\bullet$ The magnetic quantum number ($m_l = -l,-l+1,...,0,1,...,l-1,l$): relates to the angular momentum along the z-axis by $L_z = m_l\hbar$

Note: If there is a magnetic filed then this is considered the z-axis, hence the name "magnetic quantum number."

Additionally a fourth quantum number was later discovered. This number describes the intrinsic angular momentum of an electron,

$\bullet$ The intrinsic spin quantum number $m_s = +\frac{1}{2}, -\frac{1}{2}$

<h3> So, why review quantum numbers? </h3>

So, why have I spent so much time reviewing the quantum numbers? Well, together, the four quantum $(n,l,m_l,m_s)$ numbers represent the qunatum state of an electron. When electrons absorb or a emit photons, there are certain quantum states that they are more likely to transition to. Selection rules give the probabilityfor a transition state to take place. For this reason it is important to review and understand the four quantum numbers.

<h2> Selection Rules </h2>

<h3> Terminology for selection rules </h3>

$\bullet \textbf{Allowed Transition}$: An allowed transition descirbes a probable transition.

$\bullet \textbf{Forbidden Transition}$: This term is a bit confusing, forbidden transitions do happen, however they are less probable. It might be better to think of this term as "less likely transition."

<h3> What are the selection rules for the four quantum numbers? </h3>

For an allowed transition the following criteria must be met.

$\bullet$ The change in $n$ does not affect the transition probability

$\bullet$ The change is $l$ between states must be either -1 or +1

$\bullet$ The change in $m_l$ between states must be -1,0, or +1

$\bullet$ The change in $m_s$ does not affect the transition proabability

In basis, only the change in $l$ and $m_l$ determine if the transition is allowable or not.

In this diagram several allowable transitions are represented.
```python
#(1,0,0,+1/2)
plt.scatter(0,1,label = "(1,0,0,+1/2)", c = "black", marker='s', s = 200)
#(2,0,0,+1/2)
plt.scatter(0,2,label = "(2,0,0,+1/2)", c = "red", marker='s', s = 200)
#(2,1,+1,+1/2)
plt.scatter(1,2,label = "(2,1,1,+1/2)", c = "purple", marker='s', s = 200)
#(3,0,0,+1/2)
plt.scatter(0,3,label = "(3,0,0,+1/2)", c = "blue", marker='s', s = 200)
#(3,1,+1,+1/2)
plt.scatter(1,3,label = "(3,1,1,+1/2)", c = "green", marker='s', s = 200)
#(3,2,+2,+1/2)
plt.scatter(2,3,label = "(3,2,2,+1/2)", c = "maroon", marker='s', s = 200)

#(3,2,+2,+1/2) -> (3,1,+1,+1/2) and (3,2,+2,+1/2) -> (2,1,+1,+1/2)
#i made this maroon for clarity
plt.plot([1,2], [2,3], c = "maroon")
plt.plot([1,2], [3,3], c = "maroon")

#(3,1,+1,+1/2) -> (2,0,0,+1/2) and (3,1,+1,+1/2) -> (1,0,0,+1/2)
plt.plot([1,0], [3,2], c = "green")
plt.plot([1,0], [3,1], c = "green")

#(2,0,0,+1/2) -> (2,1,+1,+1/2)
plt.plot([0,1], [2,2], c = "red")

#(3,0,0,+1/2) -> (3,1,+1,+1/2) and (3,0,0,+1/2) -> (2,1,+1,+1/2)
plt.plot([0,1], [3,3], c = "blue")
plt.plot([0,1], [3,2], c = "blue")

#(2,1,+1,+1/2) -> (1,0,0,+1/2)
plt.plot([1,0], [2,1], c = "purple")

plt.xlabel("Orbital Angular Momentum Quantum Number (l)")
plt.ylabel("Principle Quantum Number (n)")
plt.title("Example diagram of allowable transition states")
plt.legend(fontsize = 10)
plt.show()
```

Notice, that the (3,2,+2,+1/2) state can transistion to (3,1,1,+1/2) and (2,1,1,+1/2) (in an allowable fashion) because in both of these transitions $\Delta(l)$ is -1 and $\Delta(m_l)$ is -1. These transtions follow the selection rules that $\Delta(m_l) = -1,0, or 1 and \Delta(l) =-1,1$. 

While the two transitions discussed are allowed, a transtion from (3,2,2,+1/2) to (1,0,0,+1/2) is forbidden. $\Delta(m_l)$ and $\Delta(l)$ are both  $2$, which by selection rules means the transition is not allowed.

<h2> Why do these selection rules exist </h2>

From earlier sections of the course we are already familiar with the fact that the photon has quantized energy of $E = \nu\hbar$. Well, it turns out that the photon also has a quantized intrinsic angular momentum. The quantization of angular momentum in a photon is $\hbar$.

By conservation of energy, the angular momentum of the atom must change by $\hbar$ when a photon is absorbed or emiited, since $\hbar$ is the intrinsic angular momentum of the photon.

So, how does this connect to selection rules?

Well, when a transition takes place a photon is either emitted or absorbed, meaning the atom either gains an $\hbar$ unit or loses an $\hbar$ unit of angular momentum. Since $l$ is the orbital (electron-proton) angular momentum number it make sense that it can only change by $|l|$ per transition because only one photon is either emitted or absorbed per transition.

$m_l$ is related to $l$ as we have discussed previously, so, in basis, these selection rules exist because of the quantization of photons and the conservation of angular momentum in the atom during transitions.
